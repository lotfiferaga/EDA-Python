```python
#Necessary Imports
import pandas as pd
import numpy as np
import seaborn as sns                       # For dataviz
import matplotlib.pyplot as plt             # For dataviz
%matplotlib inline
sns.set(color_codes=True)
```

```python
df = pd.read_csv("dataset.csv")
# To display the top 10 rows
df.head(10)
```

```python
#Display the tail or the last few rows at the end of the dataset
df.tail(10)
```

```python
#Checking all columns data types
df.dtypes
```

```python
#Dropping irrelevant columns
df = df.drop(['Engine Fuel Type', 'Market Category', 'Vehicle Style', 'Popularity', 'Number of Doors', 'Vehicle Size'], axis=1)
df.head(5)
```

```python
# Renaming columns
df = df.rename(columns={"Engine HP": "HP", "Engine Cylinders": "Cylinders", "Transmission Type": "Transmission", "Driven_Wheels": "Drive Mode","highway MPG": "MPG-H", "city mpg": "MPG-C", "MSRP": "Price" })
df.head(5)
```

```python
# Getting the dims of the dataset
df.shape
```

```python
# Counting the dataset rows
df.count()
```

```python
duplicate_rows_df = df[df.duplicated()]
print("number of duplicate rows: ", duplicate_rows_df.shape)
```

```python
# Dropping duplicates
df = df.drop_duplicates()
df.head(5)
```

```python
# Counting the rows after the drop
df.count()
```

```python
# How many row has null values
print(df.isnull().sum())
```

```python
 # Dropping the missing values
df = df.dropna()
df.count()
```

```python
# After dropping the values
print(df.isnull().sum())
```

```python
# Detecting outliers using boxplot
sns.boxplot(x=df['Price'])
```

```python
sns.boxplot(x=df['HP'])
```

```python
sns.boxplot(x=df['Cylinders'])
```

```python
# Calculating the first and the third Quantils
Quant1 = df.quantile(0.25)
Quant3 = df.quantile(0.75)
IQR = Quant3 - Quant1
print(IQR)
```

```python
df = df[~((df < (Quant1 - 1.5 * IQR)) |(df > (Quant3 + 1.5 * IQR))).any(axis=1)]
df.shape
```

**As observed earlier, approximately 1600 rows exhibited outlier values. However, it's not feasible to entirely eliminate all outliers, as even after applying the aforementioned technique, there might be 12 outliers remaining. This is acceptable, considering the initial count of over 100 outliers, as having some outliers removed is better than none.**
